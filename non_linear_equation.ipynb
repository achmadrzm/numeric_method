Mencari nilai f(x)=0 dengan metode BRACKETING (BISECTION)

```python
fx = lambda x: x**3 - 2*(x**2) + 6*x - 4.0

a = float(input("Masukkan a: "))
b = float(input("Masukkan b: "))

if (fx(a) * fx(b)) < 0:

    while abs(fx(a) * fx(b)) > 0.000001:
        c = (a + b) / 2

        if (fx(a) * fx(c)) < 0.0:
            b = c
        else:
            a = c
    print("Akar persamaan adalah %f" % (c))

else:
    print("Penentuan a dan b salah")
```

```python
import numpy as np
import matplotlib.pyplot as plt

fx = lambda x: np.sin(10*x) + np.cos(3*x)

a = float(input("Masukkan a: "))
b = float(input("Masukkan b: "))

if (fx(a) * fx(b)) < 0:

    x = np.arange(3.0, 6.0, 0.01)
    y = fx(x)

    plt.plot(x, y, '-')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(True)
    plt.axhline(y=0, color='r')
    plt.show()

    i = 0
    c_p = None

    print("Iterasi\t\t a\t b\t Nilai Tengah\t f(c)\t Relative Error (%)")
    print("-" * 65)

    while True:
        c = (a + b) / 2

        if c_p is not None:
            relative_error = abs((c - c_p) / c) * 100
        else:
            relative_error = None

        print(f"{i:<10} {a:<10.6f} {b:<10.6f} {c:<10.6f} {fx(c):<15.6f} {relative_error if relative_error is not None else 'N/A'}")

        if relative_error is not None and relative_error < 0.5:
            break

        if (fx(a) * fx(c)) < 0.0:
            b = c
        else:
            a = c

        c_p = c
        i += 1

    print("Akar persamaan adalah %f" % (c))

else:
    print("Penentuan a dan b salah")
```

```python
#Tugas: Mencari semua nilai x
#Clue: Menggunakan list, jika nilai c ditemukan, maka nilai c akan diappend kedalam list
#Contoh: Menjajaki satu per satu misal setiap 0.5. Contoh: a=1.0 b=1.5, lalu nilai c nya dimasukkan kedalam list, kemudian nilai a dan  b diupdate menjadi a=1.5 b=2.0

import numpy as np
import matplotlib.pyplot as plt

fx = lambda x: np.sin(10*x) + np.cos(3*x)

start = float(input("Masukkan batas awal: "))
end = float(input("Masukkan batas akhir: "))

roots = []
step = 0.01

def bisection(a, b):
    i = 0
    while abs(fx(a) * fx(b)) > 0.000001:
        c = (a + b) / 2
        if (fx(a) * fx(c)) < 0.0:
            b = c
        else:
            a = c
        i += 1
    return c

a = start
while a < end:
    b = a + step
    if fx(a) * fx(b) < 0:
        c = bisection(a, b)
        roots.append(c)
    a = b

if roots:
    print("Akar-akarnya adalah: ", roots)
else:
    print("Tidak ada akar yang ditemukan pada interval yang diberikan.")

x = np.arange(start, end, 0.01)
y = fx(x)
plt.plot(x, y, label='f(x) = sin(10x) + cos(3x)')
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('X')
plt.ylabel('f(X)')
plt.title('Grafik f(x)')
plt.grid(True)

for root in roots:
    plt.plot(root, 0, 'ro')

plt.show()
```

Mencari nilai f(x)=0 dengan metode BRACKETING (FALSE POSITION)

```python
fx = lambda x: x**3 - 2*(x**2) + 6*x - 4.0
a = float(input("Masukkan a: "))
b = float(input("Masukkan b: "))
cp = a
Er = 10000.0
if (fx(a) * fx(b)) < 0:
    n = 0
    print("Iterasi\t a\t b\t c\t fx\t Er")
    while abs(Er) > 0.5:
        c = b - (fx(b)*(a-b))/(fx(a)-fx(b))
        cc = c
        Er = ((cc-cp)/cc) * 100
        print("%d\t%f\t%f\t %f\t %f\t %f" % (n, a, b, c, fx(c), abs(Er)))
        cp = cc
        n += 1
        
        if (fx(a) * fx(c)) < 0.0:
            b = c
        else:
            a = c
    print("Akar persamaan adalah %f" % (c))
else:
    print("penentuan a dan b salah")
```

```python
# Case Study: Bisection

import numpy as np
import matplotlib.pyplot as plt

fx = lambda x: x**10 - 1.0

a = float(input("Masukkan a: "))
b = float(input("Masukkan b: "))

if (fx(a) * fx(b)) < 0:
    cp = a
    Er = 10000.0
    n = 0

    x = np.arange(0.9, 1.2, 0.01)
    y = fx(x)

    plt.plot(x, y, '-')
    plt.axhline(y=0, color='r', linestyle='--')
    plt.xlabel('X')
    plt.ylabel('f(X)')
    plt.title('Grafik f(x)')
    plt.grid(True)

    print("Iterasi\t a\t b\t c\t fx\t Er")
    
    while abs(Er) > 0.1:  # Galat standar 0.1% sesuai soal
        c = (a + b) / 2
        cc = c
        Er = ((cc - cp) / cc) * 100
        cp = cc
        print("%d\t%f\t%f\t %f\t %f\t %f" % (n, a, b, c, fx(c), abs(Er)))
        n += 1

        if (fx(a) * fx(c)) < 0.0:
            b = c
        else:
            a = c

    print("Akar persamaan adalah %f" % (c))

else:
    print("Penentuan a dan b salah")
```
